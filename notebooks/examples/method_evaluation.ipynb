### CoolPrompt method evaluation interface

With our coolprompt.method_evaluation pipeline you can easilly evaluate any autoprompting algorithm (you can either choose one of our methods or implement and evaluate your own!)

See more about your own method implementation at notebooks/examples/custom_apo_method.ipynb

```python
from coolprompt.method_evaluation import evaluate_method
```

```python
from langchain_openai import ChatOpenAI
import os

os.environ["OPENAI_API_KEY"] = "sk-proj-..."

model = ChatOpenAI(model="gpt-4o-mini")
```

```python
config = {
    'dataset': {
        'name': 'gsm8k',
        'configuration': '10/10/10'
    },
    'method': {
        'population_size': 3,
        'num_epochs': 2,
        'use_cache': True
    },
    'task': 'generation',
    'metric': 'em',
    'model_answers_output_path': './model_answers_examples.yaml'
}
```

```python
evaluate_method(
    model=model,
    method='hyper_light',
    config=config,
    start_prompt="Provide the solution",
    saving_model_answers=True
)
```
