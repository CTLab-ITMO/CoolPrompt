```python
from coolprompt.assistant import PromptTuner

from langchain_core.rate_limiters import InMemoryRateLimiter
from langchain_openai.chat_models import ChatOpenAI

rate_limiter = InMemoryRateLimiter(
    requests_per_second=5,
    check_every_n_seconds=0.5,
    max_bucket_size=10
)

my_model = ChatOpenAI(
    model="gpt-4o-mini",
    openai_api_key="your-api-key",
    temperature=0.1,
    max_tokens=4000,
    timeout=60,
    max_retries=10,
    rate_limiter=rate_limiter
)
```

```python
assistant = PromptTuner(my_model)
```

```python
from src.utils.load_dataset_coolprompt import load_dataset

dataset_name = "gsm8k"
dataset, target = load_dataset(dataset_name, size=40)
```

```python
start_prompt = "Find the solution"
```

```python
final_prompt = assistant.run(
    start_prompt,
    task="generation",
    dataset=dataset,
    target=target,
    method='regps',
    problem_description_generation_method='dataset-based',
    validation_size=0.5,
    metric="em", 
    output_path=f"./test_logs",
    num_epochs=2,
    population_size=5,
    bad_examples_number=7
)
final_prompt
```
